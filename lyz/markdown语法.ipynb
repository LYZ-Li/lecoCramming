标题使用#号，几个#号就是第几级， # 号之后使用一个空格与标题分开

使用空白行来做段落

换行的方式比较混乱，  
可以在句子末尾加两个或以上的空格   
或者有的语言可以直接使用回车  
当然也可以使用html语法加一个`<br>`<br>

在前后各加两个*号或者下划线，可以**粗体 bold**  
如果是加一个的话就是*斜体*  
三个星号或者下划线就是 ***粗斜体*** 

引用使用右书名号
> 这是一条引用
>>这是嵌套的引用
> - 小点是加一个横杠，加号+ 或者 一个星号*
